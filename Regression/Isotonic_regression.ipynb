```python
import random as rn
from sklearn.isotonic import IsotonicRegression
import matplotlib.pyplot as plt
```

```python
ages=[]
weight=[]
for i in range(100):
    age=rn.randint(10,70)
    ages.append(age)
    height=rn.randint(154,172)
    weight.append(age*0.5+rn.randint(2,15))
```

```python
plt.scatter(ages,weight)
```

```python
ir = IsotonicRegression()

y_ = ir.fit_transform(ages,weight)
```

```python
plt.scatter(ages,weight)
plt.scatter(ages,y_)
```

## 3D PLOT

```python
X=[]

weight=[]
for i in range(100):
    age=rn.randint(10,70)
    height=rn.randint(154,172)
    X.append([age,height])
    
    weight.append(age*0.5+height*0.23+rn.randint(2,15))
```

```python
from mpl_toolkits.mplot3d import Axes3D
Axes3D = Axes3D
fig = plt.figure(figsize=(8,8))
ax= fig.add_subplot(111,projection='3d')


xs=[i[0] for i in X]
ys=[i[1] for i in X]
zs=weight

ax.scatter(xs,ys,zs,s=50,alpha=0.6,edgecolors='w')
ax.set_xlabel('Age') 
ax.set_ylabel('Weight') 
ax.set_zlabel('Height') 
plt.show() 
```
